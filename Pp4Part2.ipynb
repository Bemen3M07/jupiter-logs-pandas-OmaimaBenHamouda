### 1. Pandas

Pandas es una biblioteca de Python utilizada para la manipulación y análisis de datos. Proporciona estructuras de datos eficientes como DataFrame y Series, que permiten manejar grandes volúmenes de datos de manera sencilla. Permite realizar tareas como la limpieza de datos, transformación, filtrado, agrupación, agregación, y visualización.


```python
import pandas as pd

# Crear un DataFrame con Pandas
data = {
    'Nombre': ['Juan', 'Ana', 'Pedro', 'Laura'],
    'Edad': [28, 22, 34, 25],
    'Ciudad': ['Madrid', 'Barcelona', 'Sevilla', 'Valencia']
}

df = pd.DataFrame(data)

# Mostrar el DataFrame
print(df)

# Filtrar datos donde la edad sea mayor de 25
mayores_de_25 = df[df['Edad'] > 25]
print(mayores_de_25)
```

### 2. Jupyter Notebook

Jupyter Notebook es una aplicación web que permite crear y compartir documentos que contienen código en vivo, ecuaciones, visualizaciones y texto explicativo. Es especialmente útil para la exploración interactiva de datos, análisis, visualización y desarrollo de proyectos en Python. En Jupyter, puedes ejecutar bloques de código de manera incremental y ver los resultados inmediatamente.

```python
# Crear un gráfico con Matplotlib en Jupyter
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de X vs Y')
plt.show()
```


### 3. reportlab

ReportLab es una biblioteca de Python para la creación de documentos PDF. Permite generar archivos PDF a partir de código Python y es muy utilizada para crear informes, gráficos, facturas y otros documentos automáticamente.

```python

from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

# Crear un archivo PDF con ReportLab
c = canvas.Canvas("pruebaReportLab.pdf", pagesize=letter)

# Escribir texto en el PDF
c.drawString(100, 750, "¡Hola, Victor! Este es un PDF generado con ReportLab.")

# Guardar el archivo PDF
c.save()
```
