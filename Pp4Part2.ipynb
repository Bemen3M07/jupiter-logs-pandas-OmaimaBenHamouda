### 1. Pandas

Pandas es una biblioteca de Python utilizada para la manipulación y análisis de datos. Proporciona estructuras de datos eficientes como DataFrame y Series, que permiten manejar grandes volúmenes de datos de manera sencilla. Permite realizar tareas como la limpieza de datos, transformación, filtrado, agrupación, agregación, y visualización.


```python
import pandas as pd

# Crear un DataFrame con Pandas
data = {
    'Nombre': ['Juan', 'Ana', 'Pedro', 'Laura'],
    'Edad': [28, 22, 34, 25],
    'Ciudad': ['Madrid', 'Barcelona', 'Sevilla', 'Valencia']
}

df = pd.DataFrame(data)

# Mostrar el DataFrame
print(df)

# Filtrar datos donde la edad sea mayor de 25
mayores_de_25 = df[df['Edad'] > 25]
print(mayores_de_25)
```

### 2. Jupyter Notebook

Jupyter Notebook es una aplicación web que permite crear y compartir documentos que contienen código en vivo, ecuaciones, visualizaciones y texto explicativo. Es especialmente útil para la exploración interactiva de datos, análisis, visualización y desarrollo de proyectos en Python. En Jupyter, puedes ejecutar bloques de código de manera incremental y ver los resultados inmediatamente.

```python
# Crear un gráfico con Matplotlib en Jupyter
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de X vs Y')
plt.show()
```


### 3. reportlab

ReportLab es una biblioteca de Python para la creación de documentos PDF. Permite generar archivos PDF a partir de código Python y es muy utilizada para crear informes, gráficos, facturas y otros documentos automáticamente.

```python

from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

# Crear un archivo PDF con ReportLab
c = canvas.Canvas("pruebaReportLab.pdf", pagesize=letter)

# Escribir texto en el PDF
c.drawString(100, 750, "¡Hola, Victor! Este es un PDF generado con ReportLab.")

# Guardar el archivo PDF
c.save()
```

# Ejercicio 5

### Media de nota final de cada alumno


```python
mediaAlumno=dataFrameAlumnos[["M01","M02","M03","M04"]].mean(axis=1)
mediaAlumno
```

### La media de la nota final de todos los alumnos en conjunto

```python
mediaFinal=mediaAlumno.mean()
print("La media de la nota final de todos los alumnos en conjunto: ",mediaFinal)
```

### Porcentajes de aprobados y suspendidos.
 ### Aprobados: 80%
    Indica un buen desempeño general del grupo. La mayoría de los estudiantes lograron un promedio suficiente (≥5) en sus  calificaciones.

### Suspendidos: 20%
    Sugiere que una pequeña parte del grupo no alcanzó el nivel necesario para aprobar. Esto podría indicar áreas en las que #### los estudiantes necesitarían más apoyo o donde los temas podrían haber sido más difíciles para ellos.

```python
aprobados=len(mediaAlumno[mediaAlumno>=5])
suspendidos=len(mediaAlumno[mediaAlumno<5])
print("Porcentaje alumnos aprobados: ",(aprobados/len(mediaAlumno))*100)
print("Porcentaje alumnos suspendidos: ",(suspendidos/len(mediaAlumno))*100)
```

### Matplotlib 
    Es una librería de Python para crear gráficos estáticos y visualizaciones de datos. Permite generar gráficos como líneas, barras, dispersión y histogramas, y personalizarlos fácilmente (colores, etiquetas, títulos, etc.). Es muy utilizada en análisis de datos para representar visualmente tendencias y patrones.
### a. Gráfico de líneas
    Crea un gráfico de líneas que representa las notas del Módulo M01 de cada alumno. Los nombres de los alumnos están en el eje X y las calificaciones en el eje Y, mostrando cómo varían las notas a lo largo de los estudiantes. 

```python
import matplotlib.pyplot as plt
x = dataFrameAlumnos["Alumne"]
y =  dataFrameAlumnos[["M01"]]
plt.figure(figsize=(14, 6))  

plt.plot(x,y)
plt.title('Notas del Módulo M01 por Alumno"')
plt.xlabel('Alumnos')
plt.ylabel('Notas')

# Mostrar el gráfico
plt.show()
```

### b. gráfico circular
    Crea un gráfico circular (pastel) que muestra el porcentaje de aprobados y suspendidos. Los datos de "Aprobados" y "Suspendidos" se representan como dos secciones del pastel, con los porcentajes mostrados dentro del gráfico. Los colores verde y rojo se usan para representar a los aprobados y suspendidos, respectivamente.

```python
import matplotlib.pyplot as plt

labels = ['Aprobados', 'Suspendidos']  

plt.figure(figsize=(5, 5))  
plt.pie([aprobados,suspendidos], labels=labels, autopct='%1.1f%%', startangle=90, colors=['#4CAF50', '#F44336'])

# Título del gráfico
plt.title('Porcentaje de Aprobados y Suspendidos')

# Mostrar el gráfico
plt.show()
```

###  C. Histogramas
    Genera 4 histogramas en una cuadrícula 2x2 para mostrar la distribución de las calificaciones de los alumnos en los módulos M01, M02, M03 y M04. Cada histograma representa la frecuencia de las notas de los alumnos en cada módulo, lo que permite observar cómo se distribuyen las calificaciones dentro de intervalos definidos (bins).

```python
plt.figure(figsize=(8, 7))

# Histograma para M01
plt.subplot(2, 2, 1)  
plt.hist(dataFrameAlumnos['M01'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma M01')
plt.xlabel('Notas M01')
plt.ylabel('Frecuencia')

# Histograma para M02
plt.subplot(2, 2, 2)  
plt.hist(dataFrameAlumnos['M02'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Histograma M02')
plt.xlabel('Notas M02')
plt.ylabel('Frecuencia')

# Histograma para M03
plt.subplot(2, 2, 3)  
plt.hist(dataFrameAlumnos['M03'], bins=10, color='lightcoral', edgecolor='black')
plt.title('Histograma M03')
plt.xlabel('Notas M03')
plt.ylabel('Frecuencia')

# Histograma para M04
plt.subplot(2, 2, 4)  
plt.hist(dataFrameAlumnos['M04'], bins=10, color='lightyellow', edgecolor='black')
plt.title('Histograma M04')
plt.xlabel('Notas M04')
plt.ylabel('Frecuencia')

plt.tight_layout()
```

### d. Gráfico de barras 
    Este gráfico de barras muestra la comparación de las calificaciones promedio de los alumnos en los diferentes módulos (M01, M02, M03, M04). Cada barra representa el promedio de las calificaciones de todos los alumnos para un módulo específico. La altura de cada barra indica el valor promedio, lo que permite ver cuál módulo tiene la calificación más alta o más baja de forma clara y visual. Es útil para identificar rápidamente en qué módulo los estudiantes tuvieron un mejor rendimiento en promedio y en cuál necesitaron más esfuerzo.


```python
promedios = dataFrameAlumnos[['M01', 'M02', 'M03', 'M04']].mean()
plt.figure(figsize=(6, 4))
plt.bar(promedios.index, promedios.values, color=['#4CAF50', '#FF9800', '#2196F3', '#F44336'])

plt.title('Promedio de Notas por Módulo')
plt.xlabel('Módulos')
plt.ylabel('Promedio de Notas')
plt.show()
```
